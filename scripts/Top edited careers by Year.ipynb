```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stats
```

```python
fn = '../input/wikipedia_edits.csv'
df = pd.read_csv(fn)
```

```python
df.head()
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df['date_year'] = df['timestamp'].dt.year
```

```python
df.head()
```

```python
len(pd.unique(df['career']))
```

```python
df_careers = df.groupby(['date_year', 'career']).size()
```

```python
df_careers = pd.DataFrame(df_careers)
```

```python
df_careers.columns = ['edits']
```

```python
df_careers.reset_index(inplace=True)
```

```python
df_careers = df_careers.sort_values(by=['date_year', 'edits'], ascending=[True, False])
```
