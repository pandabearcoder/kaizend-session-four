```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as stats
```

```python
fn = '../input/wikipedia_edits.csv'
df = pd.read_csv(fn)
```

```python
df.head()
```

### Top 5 Edited career

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df['date_year'] = df['timestamp'].dt.year
```

```python
df.head()
```

```python
df_careers = df.groupby(['year', 'career']).size()
```

```python
df_careers = pd.DataFrame(df_careers)
```

```python
df_careers.columns = ['edits']
```

```python
df_careers.reset_index(inplace=True)
```

```python
df_careers.sort_values(by=['year', 'edits'], ascending=[True, False])
```

```python
top_five_careers_per_year = df_careers.groupby('year').head(5)
```

```python
top_five_careers_per_year
```
