## Who are the top contributors for every celebrity?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
fn = '../input/wikipedia_edits.csv.zip'
df = pd.read_csv(fn)
df.head()
```

```python
fans = df.groupby(['title', 'user']).size().to_frame('edits')
fans.reset_index(inplace=True)
fans
```

```python
fans_max = fans.groupby(['title']).max()
fans_max
```

```python
fans_max.loc[fans_max['edits'].head()
```

```python
fans_max.sort_values(by="edits", ascending=False)
```

```python
# Rank by being top contributors
fans_max.groupby(['user']).size().sort_values(ascending=False)
```
