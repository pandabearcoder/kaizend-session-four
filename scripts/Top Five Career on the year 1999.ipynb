## Top Five Career on the year 1999

```python
import pandas as pd
import numpy as np

fn = '../input/forbes_celebrity_list.csv'
df = pd.read_csv(fn)
```

```python
df
```

```python
df.head()
```

```python
year = df[(df['year'] == 1999)]
sorted_v = year.sort_values(by=["year", "rank"], ascending=True)

sorted_career = sorted_v.groupby(["career"]).size().to_frame('Top 5 Careers 1999')
sorted_career.head()
```
